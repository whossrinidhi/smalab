```python
import praw
import pandas as pd

# Replace with your actual credentials
client_id = ''
client_secret = ''
user_agent = 'MyRedditApp/1.0 by yourusername'

# Create Reddit instance
reddit = praw.Reddit(
    client_id=client_id,
    client_secret=client_secret,
    user_agent=user_agent
)

# Access the subreddit
subreddit = reddit.subreddit("python")

# List to store processed post data
posts_data = []

# Fetch top 10 posts
for post in subreddit.hot(limit=100):
    # Preprocess title (e.g., remove special characters, lowercase)
    title = post.title.strip().lower()

    # Example: Filter posts based on upvotes (e.g., only posts with more than 100 upvotes)
    if post.score > 100:
        author = post.author.name if post.author else 'Deleted'
        posts_data.append({
            'Title': title,
            'Upvotes': post.score,
            'Posted By': author
        })

# Convert the list to a pandas DataFrame
df = pd.DataFrame(posts_data)

# Save to CSV
df.to_csv('reddit_posts_data.csv', index=False)

# Optionally, save to JSON
df.to_json('reddit_posts_data.json', orient='records', lines=True)

print(f"Saved {len(df)} posts data.")

csv = pd.read_csv("reddit_posts_data.csv")
csv.head()
```
